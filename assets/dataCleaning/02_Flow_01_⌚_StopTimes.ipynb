Now using `stop_lat` and `stop_lon` all stops for this direction route can be mapped out.Effectively a drop down with the stops will appear a selection must be made. It is also the time to call the API to see which buses are coming 

```python
# API Endpoint
GTFSR_API_URL = "https://api.nationaltransport.ie/gtfsr/v2/gtfsr?format=json"

# Function to fetch GTFSR data with error handling
def fetch_gtfsr_data():
    headers = {"x-api-key": YOUR_API_KEY}
    try:
        response = requests.get(GTFSR_API_URL, headers=headers, verify=False)
        response.raise_for_status()  # Raise error if status code is not 200
        data = response.json()
        
        if "entity" not in data:
            print("⚠️ 'entity' key not found in API response. Please check the API response format.")
            return None
        
        return data
    
    except requests.exceptions.HTTPError as http_err:
        print(f"HTTP Error: {http_err}")
    except requests.exceptions.RequestException as req_err:
        print(f"Request Error: {req_err}")
    except ValueError as json_err:
        print(f"JSON Decode Error: {json_err}")
    
    return None

# Fetch GTFSR data
gtfsr_data = fetch_gtfsr_data()

# Check if data was fetched
if gtfsr_data is None:
    print("❌ No data fetched. Please check the API key, network connection, or API availability.")
else:
    print("✅ Data fetched successfully!")

# Current UTC time
current_time_utc = datetime.now(pytz.utc)

# Function to process and filter GTFSR data
def process_gtfsr_data(data):
    trips = []
    
    for entity in data.get("entity", []):
        trip_info = entity.get("trip_update", {}).get("trip", {})
        stop_time_updates = entity.get("trip_update", {}).get("stop_time_update", [])

        # Extract trip details
        trip_id = trip_info.get("trip_id")
        start_time = trip_info.get("start_time")
        start_date = trip_info.get("start_date")
        route_id = trip_info.get("route_id")
        direction_id = trip_info.get("direction_id")
        schedule_relationship = trip_info.get("schedule_relationship")

        # Skip if not scheduled
        if schedule_relationship != "SCHEDULED":
            continue

        # Combine start_date and start_time for filtering
        if not start_time or not start_date:
            continue  # Skip if start_time or start_date is missing

        try:
            trip_datetime_str = f"{start_date} {start_time}"
            trip_datetime = datetime.strptime(trip_datetime_str, "%Y%m%d %H:%M:%S").replace(tzinfo=pytz.utc)
        except ValueError:
            continue  # Skip trips with invalid datetime format

        # Filter: Trips that started within the last 3 hours
        if current_time_utc - timedelta(hours=3) <= trip_datetime <= current_time_utc:
            for stop in stop_time_updates:
                stop_sequence = stop.get("stop_sequence")
                stop_id = stop.get("stop_id")
                arrival_delay = stop.get("arrival", {}).get("delay", 0)
                departure_delay = stop.get("departure", {}).get("delay", 0)

                trips.append({
                    "trip_id": trip_id,
                    "route_id": route_id,
                    "direction_id": direction_id,
                    "stop_id": stop_id
                })

    return pd.DataFrame(trips)

# Process and filter the data
if gtfsr_data:
    scheduled_time = process_gtfsr_data(gtfsr_data)

    if scheduled_time.empty:
        print("⚠️ No scheduled trips found within the last 3 hours.")
    else:
        # ✅ Create the 'concat' column
        scheduled_time['concat'] = (
            scheduled_time['route_id'].astype(str) + 
            scheduled_time['direction_id'].astype(str)
        )


        # ✅ Apply Filter : Filter by concat
        scheduled_time = scheduled_time[scheduled_time['concat'] == concat]
        
scheduled_time.head(10)
```

```python
# Get distinct trip_id values as a DataFrame
unique_trip_ids_df = scheduled_time[['trip_id']].drop_duplicates()

# Print the unique trip_ids DataFrame
print("🔎 Distinct trip_id values (DataFrame):")
print(unique_trip_ids_df)
```

Combine stop selected and unique trip_id to find out arrival times for the selected stop 8250DB005052

```python
# Read the StopTimesPerTrip.txt file
file_path = "../data/StopTimesPerTrip.txt" 
stpr_df = pd.read_csv(file_path)
stpr_df.head(20)
```

```python
# Filter stpr_df for rows where 'trip_id' is in unique_trip_ids_df and 'stop_id' is '8250DB005052'
filtered_stpr_df = stpr_df[
    (stpr_df['trip_id'].isin(unique_trip_ids_df['trip_id'])) &
    (stpr_df['stop_id'] == selected_stop)
]

# Display the filtered DataFrame
filtered_stpr_df
```

```python
# Function to handle '24:00:00' and beyond
def convert_gtfs_time(time_str):
    """Converts GTFS time beyond 24 hours into a valid datetime format."""
    try:
        h, m, s = map(int, time_str.split(':'))
        if h >= 24:
            h -= 24  # Normalize hours beyond 24
            return (datetime.combine(datetime.today(), datetime.min.time()) + timedelta(days=1, hours=h, minutes=m, seconds=s)).time()
        else:
            return datetime.strptime(time_str, "%H:%M:%S").time()
    except Exception as e:
        print(f"Error parsing time '{time_str}': {e}")
        return None

# Apply the conversion
filtered_stpr_df.loc[:, 'arrival_time'] = filtered_stpr_df['arrival_time'].apply(convert_gtfs_time)

# Get the current time
current_time = datetime.now().time()

# Filter out past buses
upcoming_buses_df = filtered_stpr_df[filtered_stpr_df['arrival_time'] > current_time].copy()

# Calculate minutes left for the next bus to arrive
def calculate_minutes_left(arrival_time):
    arrival_datetime = datetime.combine(datetime.today(), arrival_time)
    current_datetime = datetime.combine(datetime.today(), current_time)
    delta = arrival_datetime - current_datetime
    return int(delta.total_seconds() // 60)

# Apply the calculation
upcoming_buses_df['minutes_left'] = upcoming_buses_df['arrival_time'].apply(calculate_minutes_left)

# Display the result
upcoming_buses_df.reset_index(drop=True)
```
